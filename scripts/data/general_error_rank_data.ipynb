```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Import the data
df = pd.read_csv('general_error_rank_data_bench.csv')
```

## Error numerical experiment

Data filtering

```python
omega=0.1

filtered_df= df[(df["omega"]==omega)]
filtered_df=filtered_df[['eps','err','err_naive_0','err_tight_0','err_naive_1','err_tight_1']]
filtered_df['eps_self'] = filtered_df['eps']

# Define new labels for the plot
label_map = {
    'err': 'LAPACK gesdd',
    'err_naive_0': 'Distributed Naive',
    'err_tight_0': 'Distributed Tight',
    'err_naive_1': 'Incremental Naive',
    'err_tight_1': 'Incremental Tight',
    'eps_self': '$\epsilon*$'
}
```

Code

```python
fig=plt.figure(figsize=(12, 8))
fig.supxlabel('Prescribed $\epsilon*$')
fig.supylabel('Mean Frobenius Error')
ax1 = plt.subplot(2,1,1)
ax2 = plt.subplot(2,2,3)
ax3 = plt.subplot(2,2,4)

# Define styles
markers = ['o', 's', 'd', 's', 'd', '*']
colors = ['k', 'g', 'r', 'c', 'm', 'k']
linestyles = ['-', ':', ':','-.','-.', '--']

for i, col in enumerate(['err','err_naive_0','err_tight_0','err_naive_1','err_tight_1',"eps_self"]):
    marker = markers[i % len(markers)] if col != 'eps_self' else None
    ax1 = filtered_df.plot(
        x='eps',
        y=col,
        ax=ax1,
        label=label_map[col],
        marker=marker,
        color=colors[i % len(colors)],
        linestyle=linestyles[i]
    )

ax1.set_yscale('log')
ax1.set_xscale('log')
ax1.set_xlim([1e-1,1e-10])
ax1.set_xlabel(None)


for i, col in enumerate(['err','err_naive_0','err_tight_0','err_naive_1','err_tight_1',"eps_self"]):
    marker = markers[i % len(markers)] if col != 'eps_self' else None
    ax2 = filtered_df.plot(
        x='eps',
        y=col,
        ax=ax2,
        marker=marker,
        color=colors[i % len(colors)],
        linestyle=linestyles[i]
    )

ax2.set_yscale('log')
ax2.set_xscale('log')
ax2.set_xlim([1e-1,1e-3])
ax2.set_ylim([1e-4,2e-1])
ax2.set_xlabel(None)
ax2.get_legend().remove()

for i, col in enumerate(['err','err_naive_0','err_tight_0','err_naive_1','err_tight_1',"eps_self"]):
    marker = markers[i % len(markers)] if col != 'eps_self' else None
    ax3 = filtered_df.plot(
        x='eps',
        y=col,
        ax=ax3,
        marker=marker,
        color=colors[i % len(colors)],
        linestyle=linestyles[i]
    )

ax3.set_yscale('log')
ax3.set_xscale('log')
ax3.set_xlim([1e-4,1e-10])
ax3.set_ylim([1e-15,1e-14])
ax3.set_xlabel(None)
ax3.get_legend().remove()
```

## Rank revealing properties

Data filter

```python
omega=0.01

filtered_df= df[(df["omega"]==omega)]
filtered_df=filtered_df[['eps','rk','rk_naive_0','rk_tight_0','rk_naive_1','rk_tight_1']]

# Define new labels for the plot
label_map = {
    'rk': 'LAPACK gesdd',
    'rk_naive_0': 'Distributed Naive',
    'rk_tight_0': 'Distributed Tight',
    'rk_naive_1': 'Incremental Naive',
    'rk_tight_1': 'Incremental Tight',
}

# Define styles
markers = ['o', 's', 'd', 's', 'd', '*']
colors = ['k', 'g', 'r', 'c', 'm', 'k']
linestyles = ['-', ':', ':','-.','-.', '--']

ax=None

for i, col in enumerate(['rk','rk_naive_0','rk_tight_0','rk_naive_1','rk_tight_1']):
    marker = markers[i % len(markers)]
    ax = filtered_df.plot(
        x='eps',
        marker= marker,
        y=col,
        ax=ax,
        label=label_map[col],
        color=colors[i % len(colors)],
        linestyle=linestyles[i]
    )

ax.set_xscale('log')
ax.set_xlim([1e-1,1e-5])
ax.set_xlabel('Prescribed error $\epsilon*$')
ax.set_ylabel('Mean matrix rank')
```

## Control parameter behavior

```python
# Import the data
df = pd.read_csv('general_error_rank_data_bench_2.csv')
```

```python
filtered_df=df[['eps','omega','rk','err','err_tight_0','err_tight_1','rk_tight_0','rk_tight_1']]

markers = ['1', '2', '3', '4', '+']
colors = ['b','g', 'r', 'c', 'm',]
linestyles = [ ':', '-.']

fig2=plt.figure(figsize=(12, 4))
fig2.supxlabel('Prescribed $\epsilon*$')
ax4 = fig2.add_subplot(1,2,1)
ax5 = fig2.add_subplot(1,2,2)

for i, omega in enumerate(pd.unique(filtered_df['omega'])):
    plot_df = filtered_df[(filtered_df["omega"]==omega)]
    plot_df = plot_df[['eps','err_tight_1']]
    ax4 = plot_df.plot(
        x='eps',
        y='err_tight_1',
        marker=markers[i],
        ax=ax4,
        label="$\omega="+str(omega)+"$",
        color=colors[i],
        linestyle=linestyles[i%2]
    )
    
plot_df = filtered_df[(filtered_df["omega"]==0.1)]
ax4 = plot_df.plot(
        x='eps',
        y='err',
        label="LAPACK gesdd",
        ax=ax4,
        color='k',
        linestyle='-'
    )
    
ax4.set_yscale('log')
ax4.set_xscale('log')
ax4.set_xlabel(None)
ax4.set_xlim([4e-2,4e-4])
ax4.get_legend().remove()
ax4.set_ylabel("Mean Frobenius Error")

for i, omega in enumerate(pd.unique(filtered_df['omega'])):
    plot_df = filtered_df[(filtered_df["omega"]==omega)]
    plot_df = plot_df[['eps','rk_tight_1']]
    ax5 = plot_df.plot(
        x='eps',
        y='rk_tight_1',
        marker=markers[i],
        ax=ax5,
        label="$\omega="+str(omega)+"$",
        color=colors[i],
        linestyle=linestyles[i%2]
    )
    
plot_df = filtered_df[(filtered_df["omega"]==0.1)]
ax5 = plot_df.plot(
        x='eps',
        y='rk',
        label="LAPACK gesdd",
        ax=ax5,
        color='k',
        linestyle='-'
    )

ax5.set_xlim([4e-2,4e-4])    
ax5.set_xscale('log')
ax5.set_xlabel(None)
ax5.set_ylabel("Mean matrix rank")
```
