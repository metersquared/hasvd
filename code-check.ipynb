```python
import numpy as np
import numpy.linalg as la
from utils import random_matrix

n = 8
m = 8
r = 4
rng=np.random.default_rng(42)

#A = rng.random((n,m))
A = random_matrix(n,m,r,rng)
print(A)

stol=1e-12
```

```python
U,S,Vh = la.svd(A)

Smat=np.zeros((n,m))
Smat[:r,:r]=np.diag(S[:r])

#print(la.norm(A-U@Smat@Vh))
print(U)
print(S)
print(Vh)
```

## Column partitioning

Partitioning

```python
A1=A[:,:2]
A2=A[:,2:]
```

SVD of first block

```python
U1,S1,Vh1 = la.svd(A1)
r1=len(S1[S1>stol])
S1mat=np.zeros((r1,A1.shape[1]))
S1mat[:r1,:r1]=np.diag(S1[:r1])
```

SVD of second block

```python

U2,S2,Vh2 = la.svd(A2)
r2=len(S2[S2>stol])
S2mat=np.zeros((r2,A2.shape[1]))
S2mat[:r2,:r2]=np.diag(S2[:r2])
```

Merging

```python
Atil=np.concatenate((U1[:,:r1]@S1mat,U2[:,:r2]@S2mat), axis=1)
Util, Stil, Vhtil = la.svd(Atil)

Stilmat=np.zeros((r,r))
Stilmat[:r,:r]=np.diag(Stil[:r])

newVh=la.inv(Stilmat)@Util[:,:r].T@A

rnew=len(Stil[Stil>stol])

Stilmat2=np.zeros((rnew,Atil.shape[1]))
Stilmat2[:rnew,:rnew]=np.diag(Stil[:r])

newVh21=np.concatenate((Vh1,np.zeros((2,6))), axis=1)
newVh22=np.concatenate((np.zeros((6,2)),Vh2), axis=1)
newVh2=np.concatenate((newVh21,newVh22), axis=0)

#print(Util)
#print(Stil)
#print(Vhtil.shape)
#print(newVh)

print(la.norm(A-Util[:,:r]@Stilmat@newVh))
print(la.norm(A-Util[:,:r]@Stilmat2@Vhtil@newVh2))
```

## Row partitioning

Partitioning

```python
A1=A[:2,:]
A2=A[2:,:]

print(A2)
```

First block

```python
U1,S1,Vh1 = la.svd(A1)
r1=len(S1[S1>stol])
S1mat=np.zeros((A1.shape[0],r1))
S1mat[:r1,:r1]=np.diag(S1[:r1])
print(S1mat.shape)
print(U1.shape)
```

Second block

```python
U2,S2,Vh2 = la.svd(A2)
r2=len(S2[S2>stol])
S2mat=np.zeros((A2.shape[0],r2))
S2mat[:r2,:r2]=np.diag(S2[:r2])

print(S2mat.shape)
print(U2.shape)
```

Merging

```python
Atil=np.concatenate((S1mat@Vh1[:r1,:],S2mat@Vh2[:r2,:]), axis=0)
print(Atil.shape)
Util, Stil, Vhtil = la.svd(Atil)


Stilmat=np.zeros((r,r))
Stilmat[:r,:r]=np.diag(Stil[:r])

print(Stilmat)


rnew=len(Stil[Stil>stol])

Stilmat2=np.zeros((Atil.shape[0],rnew))
Stilmat2[:rnew,:rnew]=np.diag(Stil[:r])

newU21=np.concatenate((U1,np.zeros((2,6))), axis=1)
newU22=np.concatenate((np.zeros((6,2)),U2), axis=1)
newU2=np.concatenate((newU21,newU22), axis=0)

print(la.norm(A-(newU2@Util@Stilmat2@Vhtil[:r,:])))
```
